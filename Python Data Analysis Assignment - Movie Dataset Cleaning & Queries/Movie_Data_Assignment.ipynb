### **Question 36**
(Visualization) 'Drama' aur 'Action' genres ki movies ke 'Gross' revenue ko compare karne ke liye ek Grouped Box Plot ya Violin Plot banayein.

```python
drama_action = df[df['Genre'].str.contains('Drama|Action', na=False)].copy()
drama_action['Main_Genre'] = drama_action['Genre'].apply(
    lambda g: 'Drama' if 'Drama' in g else 'Action'
)

plt.figure(figsize=(8, 6))
sns.violinplot(data=drama_action, x='Main_Genre', y='Gross', hue='Main_Genre',
               palette='Set2', legend=False)
plt.title('Gross Revenue Comparison: Drama vs Action')
plt.xlabel('Genre')
plt.ylabel('Gross Revenue')
plt.tight_layout()
plt.show()
```

### **Question 40**
(Visualization) Top 5 highest rated movies aur unke 'Meta_score' vs 'IMDB_Rating' (scaled to 100) ka ek Side-by-Side Comparison Bar Chart banayein.

```python
top5_rated = df.sort_values(by='IMDB_Rating', ascending=False).head(5).copy()
top5_rated['IMDB_Scaled'] = top5_rated['IMDB_Rating'] * 10

x = range(len(top5_rated))
width = 0.35

plt.figure(figsize=(10, 6))
plt.bar([i - width/2 for i in x], top5_rated['Meta_score'], width, label='Meta Score', color='teal')
plt.bar([i + width/2 for i in x], top5_rated['IMDB_Scaled'], width, label='IMDB Rating (scaled to 100)', color='salmon')
plt.xticks(list(x), top5_rated['Series_Title'], rotation=30, ha='right')
plt.ylabel('Score')
plt.title('Top 5 Highest Rated Movies: Meta Score vs IMDB Rating (scaled)')
plt.legend()
plt.tight_layout()
plt.show()
```
